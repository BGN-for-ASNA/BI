```python
import plotly.graph_objects as go
import numpy as np
import math

# Define the two lines' endpoints
x1_line1, y1_line1 = 0, 0
x2_line1, y2_line1 = 1, 4  # Line 1 coordinates


x1_line2, y1_line2 = 0, 0
x2_line2, y2_line2 = 4.5, 3  # Line 2 coordinates

# Calculate slopes of the two lines
slope_line1 = (y2_line1 - y1_line1) / (x2_line1 - x1_line1)
slope_line2 = (y2_line2 - y1_line2) / (x2_line2 - x1_line2)

# Calculate the angles of the two lines relative to the x-axis
angle_line1 = math.degrees(math.atan(slope_line1))
angle_line2 = math.degrees(math.atan(slope_line2))

# Calculate the angle between the two lines
theta = abs(angle_line2 - angle_line1)

# Parametric equations for drawing the angle arc
# We will draw the arc between the two angles (1 and 2)
angle_start = min(angle_line1, angle_line2)
angle_end = max(angle_line1, angle_line2)

# Define a radius for the arc
radius = 1

# Parametric equation for arc (portion of circle)
theta_arc = np.linspace(math.radians(angle_start), math.radians(angle_end), 100)
arc_x = radius * np.cos(theta_arc) + x1_line1
arc_y = radius * np.sin(theta_arc) + y1_line1

# Create figure
fig = go.Figure()

# Plot line 1
fig.add_trace(go.Scatter(x=[x1_line1, x2_line1], y=[y1_line1, y2_line1], mode='lines',
                         line=dict(color='green', width=2), name="Line 1"))

# Plot line 2
fig.add_trace(go.Scatter(x=[x1_line2, x2_line2], y=[y1_line2, y2_line2], mode='lines',
                         line=dict(color='red', width=2), name="Line 2"))

# Plot the arc for the angle
fig.add_trace(go.Scatter(x=arc_x, y=arc_y, mode='lines',
                         line=dict(color='blue', width=2), name=f"Angle = {theta:.2f}"))

# Mark the intersection point
fig.add_trace(go.Scatter(x=[x1_line1], y=[y1_line1], mode='markers', marker=dict(size=10, color='black'), name="Origin"))

# Add text to display the angle value
fig.add_trace(go.Scatter(x=[x1_line1 + 0.2], y=[y1_line1 + 0.2], text=[f'{theta:.2f}'], mode='text', textposition="top right"))

# Update layout
fig.update_layout(
    title=f'Angle between Line 1 and Line 2: {theta:.2f}',
    xaxis=dict(range=[-1, 5]),
    yaxis=dict(range=[-1, 5]),
    width=700,
    height=700
)

# Show figure
fig.show()
```
