#  Random Cluster Assignment
We generate a brain with random clusters to simulate structures

```python
import jax.numpy as jnp
import jax

def rand_cl(grid_size = (10, 10), block_size = 3):
    grid_size = (10, 10)
    block_size = 3

    # Calculate number of blocks along each dimension (round up to cover the whole grid)
    num_blocks_x = (grid_size[0] + block_size - 1) // block_size
    num_blocks_y = (grid_size[1] + block_size - 1) // block_size

    # Assign random values to each block
    key = jax.random.PRNGKey(0)
    block_values = jax.random.randint(key, (num_blocks_x, num_blocks_y), 0, 10)

    # Expand the block values into a grid
    array = jnp.kron(block_values, jnp.ones((block_size, block_size)))

    # Crop to desired grid size
    array = array[:grid_size[0], :grid_size[1]]

    return array

brain = jnp.zeros((10, 10, 5)) # 10 by 10 array on 5 layers

for i in range(brain.shape[2]):
    brain = brain.at[:,:, i].set(rand_cl(brain.shape[:1]))
brain
```

```python
import jax.numpy as jnp
import jax

def rand_cl_3d(grid_size=(10, 10, 5), block_size=3,seed=0):
    # Extract each dimension size
    size_x, size_y, size_z = grid_size
    
    # Calculate number of blocks along each spatial dimension
    num_blocks_x = (size_x + block_size - 1) // block_size
    num_blocks_y = (size_y + block_size - 1) // block_size
    
    # Generate random block values for each layer
    key = jax.random.PRNGKey(seed)
    block_values = jax.random.randint(
        key, (num_blocks_x, num_blocks_y, size_z), 0, 10
    )
    
    # Expand each layer's block values
    expanded = jnp.repeat(
        jnp.repeat(block_values, block_size, axis=0), 
        block_size, axis=1
    )
    
    # Crop to the desired grid size
    array = expanded[:size_x, :size_y, :]
    
    return array

brain = rand_cl_3d()
brain
```

# Define effect of categories on brain clusters



We need create N data points by generating categories effect in privous section and only one categorie per data point

```python
Ncat=len(jnp.unique(brain))
N = 100
backup=brain
brains=[]
categories=[]
intercept= 210
coefs=bi.dist.normal(0,1,shape=(Ncat,),sample=True)
for i in range(N):
    # sample a category
    category = np.random.randint(1, Ncat)
    categories.append(category)
    tmp=brain.copy()
    tmp= tmp.at[jnp.where(tmp==category)].set(
        bi.dist.normal(intercept, 10, sample=True,shape=tmp[jnp.where(tmp==category)].shape)+
        bi.dist.normal(coefs[category-1] * i+1, 1, sample=True,shape=tmp[jnp.where(tmp==category)].shape)
        ) 
    tmp= tmp.at[jnp.where(tmp!=category)].set(
        bi.dist.normal(intercept, 10, sample=True,shape=tmp[jnp.where(tmp!=category)].shape)
        )
    brains.append(tmp)
    
brains=jnp.array(brains)
```

# Convultion

```python
import jax
import jax.numpy as jnp
import jax.nn 
```

```python
x=jnp.linspace(0, 10, 20)
window = jnp.ones(10) / 10
y=jnp.convolve(x, window, mode='same')
y
```

The convolution operation here slides the window over x and at each step computes a weighted sum. In formula form, for each output index k:

$$
y[k] = \sum_{i=0}^{N-1} windows[i] \cdot x[k-offset+i]
$$

Because our window has length 10 (an even number), many libraries (including JAXs implementation, similar to NumPys) choose an offset of 5. This means that for a given output index k, the convolution tries to use the input values from indices:

$$
k5,k5+1,,k+4.
$$


```python
y=jax.nn.relu(y)
y = y[jnp.newaxis, jnp.newaxis, :, jnp.newaxis]
y
```
