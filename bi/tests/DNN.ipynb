#  Random Cluster Assignment
We generate a brain with random clusters to simulate structures

```python
import jax.numpy as jnp
import jax

def rand_cl(grid_size = (10, 10), block_size = 3):
    grid_size = (10, 10)
    block_size = 3

    # Calculate number of blocks along each dimension (round up to cover the whole grid)
    num_blocks_x = (grid_size[0] + block_size - 1) // block_size
    num_blocks_y = (grid_size[1] + block_size - 1) // block_size

    # Assign random values to each block
    key = jax.random.PRNGKey(0)
    block_values = jax.random.randint(key, (num_blocks_x, num_blocks_y), 0, 10)

    # Expand the block values into a grid
    array = jnp.kron(block_values, jnp.ones((block_size, block_size)))

    # Crop to desired grid size
    array = array[:grid_size[0], :grid_size[1]]

    return array

brain = jnp.zeros((10, 10, 5)) # 10 by 10 array on 5 layers

for i in range(brain.shape[2]):
    brain = brain.at[:,:, i].set(rand_cl(brain.shape[:1]))
brain
```
